```python
# Basic text preprocessing with nltk 2
# importing libraries
import pandas as pd
demo=open('Downloads/demotext2.txt','r')
text=demo.read()
demo.close()
text
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# Tokenization
token = nltk.sent_tokenize(text)
token
```

```python
token_1 = nltk.word_tokenize(text)
token_1
```

```python
# POS Tagging (method-01)
for tokens in token_1:
    print("POS Tagging for word:",nltk.pos_tag([tokens]))
```

```python
# POS Tagging (method-02)
tag = nltk.pos_tag(token_1)
tag
```

```python
# stopwords Tagging
from nltk.corpus import stopwords
stop_word = stopwords.words('english')
print(stop_word)
```

```python
# removing stop words
print('With Stop words:\n',text)
print('\n')
filter = [w for w in tokens if not w in stop_word]
print('Filtered_words (without Stopwords)\n',filter)
```

```python
# Feature Extraction using Bag of Words.
```

```python
# feature extraction is a process of mapping the textual data to real vectors.
# changing sentences to 0's and 1's
# Bag of Words is the simplest method for feature extraction.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c = CountVectorizer(ngram_range = (1,1))
count_vec = c.fit_transform([text])
c.get_feature_names_out() #gives names or words of feature.
```

```python
count_vec.toarray()
```

```python
import pandas as pd 
df = pd.DataFrame(count_vec.toarray(), columns = c.get_feature_names_out())
df
```

```python
# TF-IDF : A function of frequency of term "t" appearing in document "d". denoted as df(t,d)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_vector = TfidfVectorizer(min_df=1, lowercase = True, stop_words="english")
review = [text]
tf_matrix =tf_vector.fit_transform(review)
tf_matrix
```

```python
tf_name = tf_vector.get_feature_names_out()
tf_df = pd.DataFrame(tf_matrix.toarray() , columns = tf_name)
tf_df # gives in 
```
